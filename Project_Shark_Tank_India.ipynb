# Measures of Variability

```python
Min=pd.DataFrame(num).min()
Max=pd.DataFrame(num).max()
Range=Max-Min
Variance=pd.DataFrame(num).var()
STD=pd.DataFrame(num).std()
Q_25=pd.DataFrame(num).quantile(.25)
Q_75=pd.DataFrame(num).quantile(.75)
IQR=Q_75-Q_25
IRQ_upper=(Q_75 + 1.5 * IQR)  
IQR_lower=(Q_25 - 1.5 * IQR)
Out1= (Max > IQR_lower)
Out2= (Min < IQR_lower)
#Outlier= (Out1 or Out2)
```

```python
Variability=pd.concat([Max,Min,Range,Variance,STD,Q_25,Q_75,IQR,IQR_lower,IRQ_upper], axis=1)
Variability.columns=["Max","Min","Range","Variance","STD","Q_25","Q_75","IQR","IQR_lower","IRQ_upper"]
Variability
```

# Measures of Distribution of data 

```python
Skewness=pd.DataFrame(num).skew()   #Positive skewness indicates that the distribution is skewed to the right (long tail on the right), while negative skewness indicates left skew (long tail on the left). A skewness value close to zero suggests a symmetrical distribution.
Kurtosis=pd.DataFrame(num).kurtosis() # Kurtosis measures the peakedness or flatness of the data distribution compared to a normal distribution. Positive kurtosis (leptokurtic) indicates a sharper peak and heavier tails than a normal distribution, while negative kurtosis (platykurtic) indicates a flatter peak and lighter tails.
Skewness_Kurtosis=pd.concat([Skewness,Kurtosis],axis=1)
Skewness_Kurtosis.columns=["Skewness","Kurtosis"]
Skewness_Kurtosis
```

```python
sns.kdeplot(Skewness_Kurtosis)
```

```python
plt.figure(figsize=(8, 4))
plt.xticks(rotation=90)
plt.tight_layout()
sns.boxplot(num)
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(num,s=200)
plt.legend(loc='upper left', bbox_to_anchor=(1, 1), fontsize='large')
plt.show()
```

```python
from scipy.stats import zscore
z_scores=zscore(num)

#outliers = (abs(z_scores) > 3).any(axis=1)
outliers = (abs(z_scores) > 3)

outlier_rows = num[outliers]  
 
outlier=num[outliers]  
outlier
```

```python
Outlier={}
for i in outlier.columns:
    d=outlier[i].unique()
    if pd.to_numeric(outlier[i], errors='coerce').notna().any():
        Outlier[i]=d 
Outlier 
```

```python
print('List of Outlier Columns::')
for key in Outlier:
    print(key)
```

```python
for i in Outlier:
    sns.boxplot(data=num[i])
    plt.title(f'Boxplot for {i}')  # Set title for each plot
    plt.show()  # Show the plot
```

# Q11. Find appearance of each sharks? 

```python
appear = {}
for i in df.columns:
    if set(df[i].unique()).issubset({0, 1})and i not in ("deal") and  i.split('_')[1] not in ("deal"):
        v = df[i].sum()
        appear[i] = v

shark_present=pd.DataFrame(list(appear.items()), columns=['shark_name', 'Appearance'])

shark_present
```

```python
sns.set_style("whitegrid")
plt.pie(shark_present['Appearance'], labels=shark_present['shark_name'], autopct='%1.2f%%', startangle=100)
plt.figure(figsize=(10, 10))
plt.show()
```
